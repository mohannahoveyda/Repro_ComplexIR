```python
import json
# the_path = "/home/mhoveyda/RSN_Z/SIGIRRSN/predictions/BM25/test_top20_sample0_2025-05-22_12-23_filtered_with_wikidata_and_wikipedia_metadata.jsonl"
the_path = "/home/mhoveyda1/REASON/predictions/E5/test_top20_sample0_2025-06-11_15-22_filtered_more_than_one_gold_with_wikidata_and_wikipedia_metadata.jsonl"
with open(the_path, "r") as f:
    lines = f.readlines()
lines = [json.loads(line) for line in lines]

len(lines)
```

```python
count_of_missing_ratings           = 0
count_of_missing_wikipedia_gold    = 0
count_of_missing_wikipedia_pred    = 0
count_of_queries_with_no_preds     = 0  

for line in lines:
    # --- missing relevance ratings ---
    if line["metadata"]["relevance_ratings"] is None:
        count_of_missing_ratings += 1

    # --- missing gold wikipedia (once per query) ---
    if any(doc["wikipedia_text"] is None for doc in line["gold_doc_metadata"]):
        count_of_missing_wikipedia_gold += 1

    # --- missing pred wikipedia (once per query) ---
    if any(doc["wikipedia_text"] is None for doc in line["pred_docs_metadata"]):
        count_of_missing_wikipedia_pred += 1

    # --- queries whose pred_maps are all False (i.e. 0 True’s) ---
    # line["pred_maps"] is a dict of {entity_name: bool}
    if not any(line["pred_maps"].values()):
        count_of_queries_with_no_preds += 1

# final printout
print("Missing ratings:               ", count_of_missing_ratings, "of", len(lines))
print("Queries missing gold wiki:     ", count_of_missing_wikipedia_gold, "of", len(lines))
print("Queries missing pred wiki:     ", count_of_missing_wikipedia_pred, "of", len(lines))
print("Queries with 0 True in pred_maps:", count_of_queries_with_no_preds, "of", len(lines))

count_valid = sum(
    1
    for line in lines
    if line["metadata"]["relevance_ratings"] is not None
       and any(line["pred_maps"].values())
)

print("Queries with ratings and ≥1 True in pred_maps:", count_valid, "of", len(lines))


count_w_True = sum(
    1
    for line in lines
    if any(line["pred_maps"].values())

)

print("Queries with ≥1 True in pred_maps:", count_w_True, "of", len(lines))

count_w_evidence = sum(
    1
    for line in lines
    if line["metadata"]["relevance_ratings"] is not None)
print("Queries with relevance_ratings:", count_w_evidence, "of", len(lines))
```

```python
filtered_lines = [line for line in lines if any(line["pred_maps"].values())]

# Save under the same name + "filtered based on pred_maps"
filtered_path = the_path.replace(".jsonl", "_filtered_based_on_pred_maps.jsonl")
with open(filtered_path, "w") as f:
    for line in filtered_lines:
        f.write(json.dumps(line) + "\n")
```
